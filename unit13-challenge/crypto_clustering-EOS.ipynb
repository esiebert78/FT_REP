### Data Preprocessing

```python
# Keep only necessary columns:
# 'CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply'
df_crypto.drop(columns=["Unnamed: 0"], inplace=True)
df_crypto.head(-1)
```

```python
# Keep only cryptocurrencies that are trading
df_crypto = df_crypto.loc[df_crypto['IsTrading'] == True]
df_crypto.head(-1)
```

```python
# Remove the "IsTrading" column
df_crypto.drop(columns=["IsTrading"], inplace=True)
df_crypto.head(-1)
```

```python

for column in df_crypto.columns:
    print(f"Column {column} has {df_crypto[column].isnull().sum()} null values")
```

```python
# Remove rows with at least 1 null value
df_crypto = df_crypto.dropna()
df_crypto.head(-1)
```

```python
TCMined_min = df_crypto['TotalCoinsMined'].min()
TCMined_max = df_crypto['TotalCoinsMined'].max()
print(TCMined_min)
print(TCMined_max)
```

```python
# Remove rows with cryptocurrencies having no coins mined
# Based on negative & positive data in coins mined (cell above), removing cryptocurrencies with 0 coins mined
df_crypto = df_crypto.loc[df_crypto['TotalCoinsMined'] != 0]
df_crypto.head(-1)
```

```python
# Drop rows where there are 'N/A' text values
# Completed above in prior step
```

```python
# Store the 'CoinName'column in its own DataFrame prior to dropping it from crypto_df
df_CoinName = df_crypto['CoinName'].to_frame()
df_CoinName.head(-1)
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm
df_crypto.drop(columns=["CoinName"], inplace=True)
df_crypto.head(-1)
```

```python
# Create dummy variables for text features
df_crypto_encoded = pd.get_dummies(df_crypto, columns=['Algorithm', 'ProofType'])
df_crypto_encoded.head()
```

```python
# Standardize data
crypto_scaled = StandardScaler().fit_transform(df_crypto_encoded)
print(crypto_scaled[0:1])
```

### Reducing Dimensions Using PCA

```python
# Use PCA to reduce dimensions to 3 principal components
# Initialize PCA model
pca = PCA(n_components=3)

# Get 3 principal components for the data.
crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
# Create a DataFrame with the principal components data
df_crypto_pca = pd.DataFrame(
    data=crypto_pca, columns=['Principal component 1', 'Principal component 2', 'Principal component 3']
)
df_crypto_pca.head(-1)
```

### Clustering Crytocurrencies Using K-Means

#### Find the Best Value for `k` Using the Elbow Curve

```python
inertia = []
k = list(range(1, 11))


# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_crypto_pca)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
def get_clusters(k, data):
    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Fit the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
# Four clusters appear to be the best value of k based upon Elbow Curve above
four_clusters = get_clusters(4, df_crypto_pca)
four_clusters.head(-1)
```
